### 5.Top 10 & country & Medal distribution along with gender

```python
df1 = df[['Team','Sex','Medal']]
```

```python
top_ten_country = df1['Team'].value_counts().head(10).index
medal = df1['Team'].value_counts().head(10).values
```

```python
plt.figure(figsize=(12,5))
sns.barplot(x=top_ten_country,y=medal)
plt.xticks(rotation=45)
plt.show()
```

#### 5.1 top 10 Country with highest no of gold medals

```python
df2 = df[['Team','Medal','Sex']]
gold = df2[df[['Team','Medal','Sex']]['Medal'] == 'Gold']
x1 =gold['Team'].value_counts().head(10).index
y1 = gold['Team'].value_counts().head(10).values
```

```python
plt.bar(x=x1,height=y1,color='gold')
plt.xticks(rotation=45)
plt.show()
```

#### 5.2 top 10 Country with highest no of silver medals

```python
df3 = df[['Team','Medal','Sex']]
silver = df3[df[['Team','Medal','Sex']]['Medal'] == 'Silver']
x2 =silver['Team'].value_counts().head(10).index
y2 = silver['Team'].value_counts().head(10).values
```

```python
plt.bar(x=x2,height=y2,color='silver')
plt.xticks(rotation=45)
plt.show()
```

#### 5.3 top 10 Country with highest no of bronze medals

```python
df4 = df[['Team','Medal','Sex']]
bronze = df4[df[['Team','Medal','Sex']]['Medal'] == 'Bronze']
x3 =bronze['Team'].value_counts().head(10).index
y3 = bronze['Team'].value_counts().head(10).values
```

```python
plt.bar(x=x3,height=y3,color='brown')
plt.xticks(rotation=45)
plt.show()
```

#### 5.4 all medal graph in one  figure

```python
fig = plt.figure(figsize=(25,7))

# Needed to add spacing between 1st and 2nd row
# Add a margin between the main title and sub-plots
fig.subplots_adjust(hspace=0.42, top=0.85)

# Add the main title
fig.suptitle("Top 10 Country & their Medals", fontsize=29)

# Add the subplots
ax1 = fig.add_subplot(1,4,1)
ax2 = fig.add_subplot(1,4,2)
ax3 = fig.add_subplot(1,4,3)
ax4 = fig.add_subplot(1,4,4)

# Add the text for each subplot
ax1.title.set_text("All Medals")
ax2.title.set_text("Gold")
ax3.title.set_text("Silver")
ax4.title.set_text("Bronze")

#filling the graphs
plt.subplot(1,4,1)
sns.barplot(x=top_ten_country,y=medal)
plt.xticks(rotation=45)

plt.subplot(1,4,2)
plt.bar(x=x1,height=y1,color='gold')
plt.xticks(rotation=45)

plt.subplot(1,4,3)
plt.bar(x=x2,height=y2,color='silver')
plt.xticks(rotation=45)

plt.subplot(1,4,4)
plt.bar(x=x3,height=y3,color='brown')
plt.xticks(rotation=45)
plt.show()
```
